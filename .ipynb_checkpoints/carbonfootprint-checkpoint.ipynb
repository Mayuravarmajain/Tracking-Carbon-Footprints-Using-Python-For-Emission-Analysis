# Synthetic Carbon Emission Data Generation & Preprocessing

```python
import pandas as pd
import numpy as np

def generate_carbon_emission_data(output_file="India_Carbon_Emissions.csv"):
    states = [
        "Andhra Pradesh", "Arunachal Pradesh", "Assam", "Bihar", "Chhattisgarh",
        "Goa", "Gujarat", "Haryana", "Himachal Pradesh", "Jharkhand", "Karnataka",
        "Kerala", "Madhya Pradesh", "Maharashtra", "Manipur", "Meghalaya",
        "Mizoram", "Nagaland", "Odisha", "Punjab", "Rajasthan", "Sikkim",
        "Tamil Nadu", "Telangana", "Tripura", "Uttar Pradesh", "Uttarakhand",
        "West Bengal"
    ]
    
    years = list(range(2010, 2025))  # 15 years of data
    sectors = ["Energy", "Industry", "Transportation", "Agriculture", "Waste Management"]
    carbon_types = ["CO", "CO2", "CH4"]
    
    # Emission factors (based on sector and carbon type)
    emission_factors = {
        "Energy": {"CO": 2.5, "CO2": 50.0, "CH4": 3.0},
        "Industry": {"CO": 1.8, "CO2": 40.0, "CH4": 2.5},
        "Transportation": {"CO": 3.0, "CO2": 30.0, "CH4": 1.0},
        "Agriculture": {"CO": 0.5, "CO2": 10.0, "CH4": 8.0},
        "Waste Management": {"CO": 0.3, "CO2": 5.0, "CH4": 6.5},
    }
    
    # Generate dataset
    data = []
    for state in states:
        for year in years:
            for sector in sectors:
                for carbon in carbon_types:
                    emission_factor = emission_factors[sector][carbon]
                    emission_value = np.random.uniform(0.5, 1.5) * emission_factor  # Add variability
                    data.append([state, year, sector, carbon, round(emission_value, 2)])
    
    # Create DataFrame
    df = pd.DataFrame(data, columns=["State", "Year", "Sector", "Carbon_Type", "Emission (Metric Tons)"])
    
    # Save to CSV
    df.to_csv(output_file, index=False)
    print(f"Dataset saved to {output_file}")

generate_carbon_emission_data()
```

# Data Preprocessing: Cleaning and Standardizing Carbon Emission Dataset

```python
# Data Preprocessing Function
def preprocess_carbon_data(input_file="India_Carbon_Emissions.csv", output_file="Cleaned_Carbon_Emissions.csv"):
    df = pd.read_csv(input_file)
    
    # Check for missing values
    if df.isnull().sum().sum() > 0:
        print("Missing values found. Filling with mean values.")
        df.fillna(df.mean(), inplace=True)
    
    # Remove duplicate rows
    df.drop_duplicates(inplace=True)
    
    # Ensure correct data types
    df["Year"] = df["Year"].astype(int)
    df["Emission (Metric Tons)"] = df["Emission (Metric Tons)"].astype(float)
    
    # Standardize column names
    df.columns = df.columns.str.strip().str.lower().str.replace(" ", "_")
    
    # Save cleaned dataset
    df.to_csv(output_file, index=False)
    print(f"Cleaned dataset saved to {output_file}")

# Run the functions to preprocess the dataset
preprocess_carbon_data()
```

# Emission Analysis: Calculating Total Carbon Emissions by State and Sector

```python
# Emission Calculation Function
def calculate_total_emissions(input_file="Cleaned_Carbon_Emissions.csv", output_file="Total_Emissions.csv"):
    df = pd.read_csv(input_file)
    
    # Aggregate total emissions per state and sector
    total_emissions = df.groupby(["state", "year", "sector"]) \
                       .agg({"emission_(metric_tons)": "sum"}) \
                       .reset_index()
    
    # Save aggregated data
    total_emissions.to_csv(output_file, index=False)
    print(f"Total emissions data saved to {output_file}")

# Run the functions to calculate emissions
calculate_total_emissions()
```

# Visualizing Carbon Emissions: Trends, Distributions, and Insights

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Visualization Functions
def plot_emission_trends(input_file="Total_Emissions.csv"):
    df = pd.read_csv(input_file)
    plt.figure(figsize=(12, 6))
    sns.lineplot(data=df, x="year", y="emission_(metric_tons)", hue="sector")
    plt.title("Emission Trends Over Time")
    plt.xlabel("Year")
    plt.ylabel("Total Emissions (Metric Tons)")
    plt.legend(title="Sector")
    plt.grid()
    plt.show()

def plot_sector_distribution(input_file="Total_Emissions.csv"):
    df = pd.read_csv(input_file)
    plt.figure(figsize=(10, 6))
    df.groupby("sector")["emission_(metric_tons)"].sum().plot(kind="pie", autopct="%.1f%%")
    plt.title("Emission Distribution by Sector")
    plt.ylabel("")
    plt.show()

def plot_emission_by_state(input_file="Total_Emissions.csv", year=2020):
    df = pd.read_csv(input_file)
    df_year = df[df["year"] == year]
    plt.figure(figsize=(12, 6))
    sns.barplot(data=df_year, x="state", y="emission_(metric_tons)")
    plt.xticks(rotation=90)
    plt.title(f"Total Emissions by State in {year}")
    plt.xlabel("State")
    plt.ylabel("Total Emissions (Metric Tons)")
    plt.show()

def plot_heatmap(input_file="Total_Emissions.csv"):
    df = pd.read_csv(input_file)
    pivot_table = df.pivot_table(index="year", columns="sector", values="emission_(metric_tons)")
    plt.figure(figsize=(10, 6))
    sns.heatmap(pivot_table, cmap="coolwarm", annot=True, fmt=".1f")
    plt.title("Heatmap of Emissions Over Time by Sector")
    plt.show()

def plot_boxplot(input_file="Total_Emissions.csv"):
    df = pd.read_csv(input_file)
    plt.figure(figsize=(12, 6))
    sns.boxplot(data=df, x="sector", y="emission_(metric_tons)")
    plt.title("Emission Distribution by Sector")
    plt.xticks(rotation=45)
    plt.show()

# Run the functions
plot_emission_trends()
plot_sector_distribution()
plot_emission_by_state()
plot_heatmap()
plot_boxplot()
```

# Per Capita Carbon Emissions: State-Wise Comparison Across CO, CO₂, and CH₄

```python
def plot_per_capita_emissions(input_file="Cleaned_Carbon_Emissions.csv"):
    df = pd.read_csv(input_file)
    
    plt.figure(figsize=(12, 6))
    sns.barplot(data=df, x="state", y="emission_(metric_tons)", hue="carbon_type")
    plt.xticks(rotation=90)
    plt.title("Per Capita Emissions for CO, CO2, and CH4 by State")
    plt.xlabel("State")
    plt.ylabel("Per Capita Emissions (Metric Tons)")
    plt.legend(title="Carbon Type")
    plt.show()

plot_per_capita_emissions()
```
